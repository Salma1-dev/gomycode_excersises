```python
import pandas as pd

food_df = pd.read_csv (r'C:\Users\pc\Documents\Market_Basket_Optimisation.csv', sep = ',')
```

```python
column_names = food_df.columns
for column in column_names:
    print(column + ' - ' + str(food_df[column].isnull().sum()))
```

```python
food_dataset = food_df.values.tolist()
big = []
for tx in food_dataset:
    small = []
    for i in tx:
        if type(i) !=str:
            continue
        else:
            small.append(i)
    big.append(small)

te=TransactionEncoder()
te_ary=te.fit(big).transform(big)    #Apply one-hot-encoding on our dataset
cleaned_tx_df=pd.DataFrame(te_ary, columns=te.columns_)
```

```python
cleaned_tx_df.isnull().sum()
```
